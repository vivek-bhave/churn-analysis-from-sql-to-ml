```python
query = """
SELECT AVG(Age) as avg_age,AVG(Tenure) as avg_Ten,AVG(Usage_Frequency) as avg_Fre,AVG(Support_Calls) as avg_SC,AVG(Payment_Delay) as avg_PD,
AVG(Total_Spend) as avg_TS,AVG(Last_Interaction) as avg_LI FROM churn ;
"""
df = pd.read_sql(query,engine)
df
```

```python
query = """
SELECT 
    Churn,
    AVG(Tenure),
    AVG(Usage_Frequency),
    AVG(Support_Calls),
    AVG(Payment_Delay),
    AVG(Total_Spend),
    AVG(Last_Interaction)
FROM churn
GROUP BY Churn;
"""
df = pd.read_sql(query,engine)
df
```

As, We can see there is not significant difference between churn 0 and 1 for Tenure and Usage Frequency.\
 But there large difference and noticeable increase for churned customer for Support Calls,Payment Delay and Total Spend and little difference for last interaction.\
 That means the stronger association of Support Calls,Payment Delay ,Total Spend and last interaction with the churned customers than Tenure and Usage Frequency.

```python
query = """
SELECT Tenure, COUNT(*)
FROM churn
GROUP BY Tenure;
"""
df = pd.read_sql(query,engine)
print(df.head(10))
print(df.tail(10))
```

Overall as we can see the customers are more if the Tenure is high which suggests the imbalance.\
So,High Tenure customers may dominate the results.

```python
query = """
SELECT Contract_Length, COUNT(*)
FROM churn
GROUP BY Contract_Length;
"""
df = pd.read_sql(query,engine)
df
```

The Annual and Quarterly contract length Customers are nearly equal to ~ 198k .\
Where, Monthly Contract Length Customers are only ~ 110k.\
Again it suggests that Annual and Quarterly Contract length customers may dominate the results due imbalance.
